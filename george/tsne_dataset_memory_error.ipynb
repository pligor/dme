## process TSNE

```python
#12 min gia 20000 => 12*9 = ~2 hours
```

```python
def TSNEprocessing(n_components):
    assert n_components >= 2
    
    def processor(inputs):
        return TSNE(n_components=n_components, perplexity=30, random_state=random_state).fit_transform(inputs)

    df_lowdim = processSeparately(inputs=XX, targets=yy, processor=processor,
                                  n_components=n_components, n_splits=9)
    
    XX_lowdim = df_lowdim.drop(labels=[target_col], axis=1)
    yy_lowdim = df_lowdim[target_col]
    
    pd.concat((XX_lowdim, yy_lowdim), axis=1).to_csv(
        os.path.realpath(os.path.join(os.getcwd(), '../Data', 'tsne_%d_components.csv' % n_components)),
        index=False)
    
    return XX_lowdim, yy_lowdim
```
